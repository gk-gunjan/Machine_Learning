```python
import numpy an np
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
from sklearn import datasets
```

```python
boston=datasets.load_boston()
boston
```

```python
x=boston.data
y=boston.target
```

```python
x.shape
```

```python
y.shape
```

```python
import pandas as pd
df=pd.DataFrame(x)
print(boston.feature_names)
df.columns=boston.feature_names
df.describe()
```

```python
df['AGE_AGE']=df.AGE**2
df['RM_AGE_TAX']=df['RM']*df['AGE']*df['TAX']
df
```

```python
x2=df.values
x3=df.values
x2.shape ,x3.shape
```

```python
from sklearn import model_selection
x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y,random_state=0)
x2_train,x2_test,y2_train,y2_test=model_selection.train_test_split(x2,y,random_state=0)
x3_train,x3_test,y3_train,y3_test=model_selection.train_test_split(x3,y,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
alg1=LinearRegression()
alg2=LinearRegression()
alg3=LinearRegression()
```

```python
alg1.fit(x_train,y_train)
alg2.fit(x2_train,y2_train)
alg3.fit(x3_train,y3_train)
```

```python
predict=alg1.predict(x_test)
predict2=alg2.predict(x2_test)
predict3=alg3.predict(x3_test)
```

```python
#without adding extra column
train_score=alg1.score(x_train,y_train)
test_score=alg1.score(x_test,y_test)
print('Train Score ',train_score)
print('Test Score ',test_score)
print('-------------------------------')
#adding extra column
train2_score=alg2.score(x2_train,y2_train)
test2_score=alg2.score(x2_test,y2_test)
print('Train 2 Score ',train2_score)
print('Test 2 Score ',test2_score)

print('-------------------------------')
#adding extra column
train3_score=alg3.score(x3_train,y3_train)
test3_score=alg3.score(x3_test,y3_test)
print('Train 3 Score ',train3_score)
print('Test 3 Score ',test3_score)
```
