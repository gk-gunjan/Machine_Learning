```python
import numpy as np
import pandas as pd
from sklearn import model_selection
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.naive_bayes import MultinomialNB
import matplotlib.pyplot as plt
import os
import re
```

```python
#get common stopwords

stopwords = {"i", "me","my","myself", "we","our","ours","ourselves","you","your","yours","yourself","yourselves","he","him",
             "his","himself","she","her","hers","herself","it","its","itself","they","them","their","theirs","themselves",
             "what","which","who","whom","this","that","these","those","am","is","are","was","were","be","been","being","have",
             "has","had","having","do","does","did","doing","a","an","the","and","but","if","or","because","as","until","while",
             "of","at","by","for","with","about","against","between","into","through","during","before","after","above","below",
             "to","from","up","down","in","out","on","off","over","under","again","further","then","once","here","there","when",
             "where","why","how","all","any","both","each","few","more","most","other","some","such","no","nor","not","only",
             "own","same","so","than","too","very","s","t","can","will","just","don","should","now"}
```

```python
#read the data

X = []
Y = []

subDirectories = os.listdir("20_newsgroups")

for className in subDirectories:
    
    all_files = os.listdir("20_newsgroups/" + className)
    
    for file in all_files:
        
        with open("20_newsgroups/" + className + "/" + file, 'r') as f:
            text = f.read()
            X.append(text)
            Y.append(className)
```

```python
#split the data into training and testing
x_train, x_test, y_train, y_test = model_selection.train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
#create vocabulary from the x_train

def createVocabulary(x_train):

    vocabulary = {}

    for doc in x_train:

        doc = re.split(r'\W+', doc)

        for word in doc:
            word = word.lower()
            if word.isalpha() and (word not in stopwords) and len(word) >= 3:
                vocabulary[word] = vocabulary.get(word, 0) + 1
                
    return vocabulary

vocabulary = createVocabulary(x_train)
    
```

```python
#sort the vocabulary on the basis of frequency of every word in descending order

vocabulary = sorted(list(vocabulary.items()), key = lambda x : x[1], reverse = True)
vocabulary
```

```python
def convertInto2d(data):

    dataset = []

    #loop through every document
    for doc in data:

        docData = []

        all_words = re.split(r'\W+', doc)
        docFreq = {}

        #build the frequency dict for the current doc
        for word in all_words:
            word = word.lower()
            if word.isalpha() and word not in stopwords:
                docFreq[word] = docFreq.get(word, 0) + 1


        #go through every feature (word) in vocabulary
        #and find its count in the current doc
        for word, freq in vocabulary:
            docData.append(docFreq.get(word, 0))

        dataset.append(docData)
        
    dataset = np.array(dataset, dtype = int)
    return dataset

x_train = convertInto2d(x_train)
x_test = convertInto2d(x_test)
```

# MultinomialNB from scratch

```python
def fit(x_train, y_train):
    
    result = {}
    all_classes = set(y_train)
    result['total_data'] = len(y_train)
    
    for current_class in all_classes:
        
        result[current_class] = {}
        
        x_train_current = x_train[y_train == current_class]
        y_train_current = y_train[y_train == current_class]
        
        #storing the total number of documents in a class
        result[current_class]['total_count'] = len(y_train_current)
        
        total_words = 0
        
        #loop through every feature
        for j in range(1500):
            
            #store the count of current word for current class
            result[current_class][j] = x_train_current[:, j].sum()
            
            total_words += result[current_class][j]
           
        #store the total number of words for the current class
        result[current_class]['total_words'] = total_words
            
    return result
            
```

```python
def probability(dictionary, x, current_class):
    
    #we changed to np.log rather than simple numbers
    #Logarithmic probabilities are used to avoid getting multiplication of actual probabilities to zero.
    output = np.log(dictionary[current_class]['total_count']) - np.log(dictionary['total_data'])
    
    for j in range(1500):
        
        #if the current word is not in the
        #test document, do not consider its
        #probability
        if x[j] == 0:
            continue
            
        #apply laplace correction
        count_of_word_in_current_class = dictionary[current_class][j] + 1
        count_of_all_word_in_current_class = dictionary[current_class]['total_words'] + 1500
        
        current_word_probability = np.log(count_of_word_in_current_class) - np.log(count_of_all_word_in_current_class)
        
        output += current_word_probability
        
    return output
    
```

```python
def predictSinglePoint(dictionary, x):
    
    all_classes = dictionary.keys()
    
    best_class = None
    best_prob = -10000
    
    for current_class in all_classes:
        
        if current_class == 'total_data':
            continue
            
        #find the probability that x belongs to the current class
        current_class_prob = probability(dictionary, x, current_class)
        
        #if the probability for current class is 
        #greater than the best_class, then update
        #the best_class and best probability
        if current_class_prob > best_prob:
            best_prob = current_class_prob
            best_class = current_class
            
    return best_class
```

```python
def predict(dictionary, x_test):
    
    y_pred = []
    
    for x in x_test:
        
        x_class = predictSinglePoint(dictionary, x)
        y_pred.append(x_class)
        
    return y_pred
```

```python
dictionary = fit(x_train, y_train)
```
