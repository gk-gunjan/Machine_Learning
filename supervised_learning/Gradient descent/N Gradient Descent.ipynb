# Gradient Descent

```python
def step_gradient(data,learning_rate,m,c):
    m_slope=0
    c_slope=0
    M=len(data)
    for i in range(M):
        x=data[i,0]
        y=data[i,1]
        m_slope+=(-2/M)*(y-m*x-c)*x
        c_slope+=(-2/M)*(y-m*x-c)
        new_m=m-learning_rate*m_slope
        new_c=c-learning_rate*c_slope
    return new_m,new_c
```

```python
def cost(m,c,data):
    total_cost=0
    N=len(data)
    for i in range(N):
        x=data[i,0]
        y=data[i,1]
        total_cost+=(1/N)*(y-m*x-c)**2
    return total_cost
```

```python
def gd(data,no_iterations,learning_rate):
    m=0
    c=0
    for i in range(no_iterations):
        m ,c =step_gradient(data,learning_rate,m,c)
        print(i," ",cost(m,c,data))
    return m ,c
```

```python
import numpy as np
def run():
    data=np.loadtxt('data.csv',delimiter=',')
    no_iterations=100
    learning_rate=0.0001
    m,c=gd(data,no_iterations,learning_rate)
    print(m,c)
```

```python
run()
```

```python
def cost(X,Y,m):
    cost_=0
    M=len(X)
    for i in range(M):
        cost_+=(1/M)*(Y[i]-sum(m*X[i]))**2
    print(cost_)
```

```python
def step_gradient(X,Y,learning_rate,m):
    m_slope=np.array([0 for i in range(len(X[0]))])
    M=len(X)
    for i in range(M):
        x=X[i]
        y=Y[i]
        for j in range(len(m_slope)):
            m_slope[j]+=(-2/M)*(y-((m*x).sum()))*x[j]
    new_m=m-learning_rate*m_slope
    cost(X,Y,new_m)
    return new_m
        
```

```python
def gd(X,Y,learning_rate,iteration):
    m=np.array([0 for i in range(len(X[0]))])
    for i in range(iteration):
        m=step_gradient(X,Y,learning_rate,m)
    return m   
```

```python
def run():
    X=data.data
    Y=data.target
    learning_rate=0.000001
    iteration=100
    m=gd(X,Y,learning_rate,iteration)
    
```

```python
run()
```
