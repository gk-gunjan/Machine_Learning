## Titanic Model

```python
import numpy as np
import pandas as pd
from sklearn import model_selection, preprocessing
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv('training_titanic_x_y_train-Copy1.csv')
data
```

```python
Y = np.array(data['Survived'])
```

```python
data
```

```python
data['Age'].fillna(value=data.Age.mean(), inplace=True)
data['Fare'].fillna(value=data.Fare.mean(), inplace=True)
data['Embarked'].fillna(value=(data.Embarked.value_counts().idxmax()), inplace=True)
```

```python
data.drop('Name', axis=1, inplace=True)
data.drop('Cabin', axis=1, inplace=True)
data.drop('Ticket', axis=1, inplace=True)
```

```python
data
```

```python
data['Sex'].unique()
```

```python
data['Sex'].replace(['male', 'female'], [1,0], inplace = True)
```

```python
data['Embarked'].unique()
```

```python
data['Embarked'].replace(['S', 'C', 'Q'], [0, 1, 2], inplace = True)
```

```python
data
```

```python
algo = LogisticRegression()
```

```python
test = pd.read_csv('test_titanic_x_test-Copy1.csv')
```

```python
test
```

```python
test.drop('Name', axis = 1, inplace = True)
test.drop('Ticket', axis = 1, inplace = True)
test.drop('Cabin', axis = 1, inplace = True)
test['Age'].fillna(value=data.Age.mean(), inplace=True)
test['Fare'].fillna(value=data.Fare.mean(), inplace=True)
test['Embarked'].fillna(value=(data.Embarked.value_counts().idxmax()), inplace=True)
```

```python
test['Sex'].replace(['male', 'female'], [1,0], inplace = True)
test['Embarked'].replace(['S','C','Q'], [0,1,2], inplace = True)
```

```python
Y = np.array(data['Survived'])
X = np.array(data.iloc[:,:-1])
```

```python
test = np.array(test)
```

```python
# Splitting the data
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y)
```

```python
# feature scaling
scaler = preprocessing.StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
test = scaler.transform(test)
```

```python
algo.fit(X_train, Y_train)
```

```python
algo.predict(X_test)
```

```python
algo.score(X_test, Y_test)
```

```python
algo.score(X_train, Y_train)
```

```python
ans = algo.predict(test)
```

```python
np.savetxt('prediction.csv', ans,fmt='%i', delimiter=",")
```
